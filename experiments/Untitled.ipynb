```python
def preprocess_text(text):
    text=re.sub(r'[^\w\s]','',text)
    tokens=text.lower()
    tokens=re.split(r'[\s_]',tokens)
    return '_'.join(tokens)
```

```python
def get_context(tokens,start,end,window_size=10):
    context=tokens[max(0,start-window_size):start]+tokens[end:end+window_size]
    return context
```

```python
def prepare_data(folder,text_file,entity_file):
#     text_df=pd.read_csv(os.path.join(folder,text_file))
#     entity_df=pd.read_csv(os.path.join(folder,entity_file))
    
    textID2entity=defaultdict(list)
    text_id_set=set()
    for i in range(entity_df.shape[0]):
        print(i)
        text_id=entity_df.iloc[i]['text_id']
        textID2entity[text_id].append(preprocess_text(entity_df.iloc[i]['entities']))
        text_id_set.add(text_id)
        if i>=1000: break
            
    data=[]
    for id in text_id_set:
        print(id)
        text=text_df.loc[text_df['text_id']==id]['text'].values[0]
        
        tokens=[]
        doc=nlp(text)
        for token in doc: tokens.append(token)
            
        entities=[]
        for ent in doc.ents:
            if ent.label_ in ['DATE', 'TIME', 'PERCENT', 'MONEY', 'QUANTITY', 'ORDINAL', 'CARDINAL']: continue
            entities.append([ent.text,ent.start,ent.end])
        
        target_entities=textID2entity[id]
        
        for entity in entities:
            ent_text=entity[0]
            results=predict(ent_text,kdwd_entities,permutations,5,forest)
            batch={}
            batch['negative']=[]
            if results:
                for res in results:
                    if res in target_entities:
                        context=get_context(tokens,entity[1],entity[2])
                        batch['pos']=res
                        batch['context']=context
                    else:
                        batch['negative'].append(res)
            if 'context' in batch: data.append(batch)
    
    return data
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
folder='./models/weights/glove'
file='glove.6B.50d.txt'
embedding_dict={}
with open(os.path.join(folder,file),'r',encoding='utf8') as f:
    for line in f:
        values=line.split()
        word=values[0]
        vector=np.array(values[1:],'float32')
        embedding_dict[word]=vector
```

```python
with open('./models/OpenKE/ent_embeddings.pkl','rb') as f:
    ent_embeddings=pickle.load(f)
    
with open('./models/OpenKE/rel_embeddings.pkl','rb') as f:
    rel_embeddings=pickle.load(f)
```

```python
X=[]
for ele in data:
    embeds=[]
    text=ele['context']
    for word in text:
        try:
            token=word.text.lower()
            token=lemmatizer.lemmatize(token)
            embed=embedding_dict[token]
            embeds.append(embed)
        except:
            embed=np.random.random(50)
            embeds.append(embed)
    
    
    if len(embeds)<2*window_size: 
        for i in range(2*window_size-len(embeds)):
            embeds.append(np.zeros(50))
    embeds=np.array(embeds)
    
    pos_ent=ele['pos']
    pos_idx=kdwd_entities.index(pos_ent)
    pos_ent_embed=ent_embeddings(torch.LongTensor([pos_idx]))
    X.append((embeds,pos_ent_embed,1))
    
    for neg_ent in ele['negative']:
        neg_idx=kdwd_entities.index(neg_ent)
        neg_ent_embed=ent_embeddings(torch.LongTensor([neg_idx]))
        X.append((embeds,neg_ent_embed,0))
    
X=np.array(X)
```

```python
def build_KG_context_model(text_input_shape, graph_input_shape, embed_dim, dict_size):
    inp_0 = Input(shape=(text_input_shape,embed_dim,))
    inp_1 = Input(shape=(graph_input_shape,))
    out_0 = Bidirectional(LSTM(units=128,return_sequences=False))(inp_0)
    w = concatenate([out_0, inp_1])
    w = Dense(256, activation='relu')(w)
    w = Dropout(0.2)(w)
    out = Dense(2, activation='sigmoid')(w)
    model = Model(inputs=[inp_0, inp_1], outputs=out)
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
text_input_shape=20
graph_input_shape=200
dict_size=len(embedding_dict)
embed_dim=50

model=build_KG_context_model(text_input_shape,graph_input_shape,embed_dim,dict_size)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
lstm_train=[]
graph_train=[]
y_train=[]

for x in X:
    lstm_train.append(x[0])
    graph_train.append(x[1].squeeze(0).detach().numpy())
    if x[2]:
        y_train.append([0,1])
    else: 
        y_train.append([1,0])

lstm_train=np.array(lstm_train)
graph_train=np.array(graph_train)
y_train=np.array(y_train)
```

```python
model.fit([lstm_train,graph_train],y_train,epochs=10)
```
